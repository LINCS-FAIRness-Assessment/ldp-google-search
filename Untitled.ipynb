```python
from google import search
import pandas as pd
from urlparse import urljoin
from collections import Counter
from clustergrammer_widget import *
```

```python
termList = pd.read_table('/Users/maayanlab/Desktop/search_terms.txt', header=None)[0].tolist()
searchCounts = {term: Counter([urljoin(url, '/') for url in search(term, num=11, start=0, stop=10)]) for term in termList}
results = pd.DataFrame(searchCounts).T.fillna(0)
lincs_results = results[[x for x in results.columns if 'lincs' in x]]
lincs_results
```

```python
# Clustergram
net = Network(clustergrammer_widget)
net.load_df(results)
net.cluster()
net.widget()
```
